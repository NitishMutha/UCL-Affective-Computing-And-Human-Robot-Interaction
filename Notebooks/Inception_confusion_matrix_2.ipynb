```python
import matplotlib
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
res = pd.read_csv('nitish_results.csv')
```

```python
res.head()
```

```python
true = res.image.map(lambda x: 1 if x.find('fear') > -1
             else 2 if x.find('happy') > -1
              else 3 if x.find('neutral') > -1 
              else 0 if x.find('anger') > -1
              else 4 )
true_val = true.values.tolist()
```

```python
pred = res[['anger','fear','happy','neutral','sad']]
pl = pred.values.tolist()
prediction = np.argmax(pl,axis=1)
prediction
```

```python
def save_confusion_matrix(title, classes, true, pred, cmap=plt.cm.Blues,normalize=True):

    # Get the confusion matrix using sklearn.
    cm = confusion_matrix(y_true=true,
                          y_pred=pred)

    # Print the confusion matrix as text.
    #print("confusion matrix: \n", cm)
    im=plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = np.round(cm.astype('float') / cm.sum(axis=1)[:, np.newaxis],2)
        #print("Normalized confusion matrix")
    else:
        pass
        #print('Confusion matrix, without normalization')

    #print(cm)

    thresh = cm.max() / 1.1
    #print(thresh)
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.colorbar(im, fraction=0.046, pad=0.04)
    
    #plt.savefig(title + '_metrics.eps', bbox_inches='tight', format='eps', dpi=50)
    #plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
import itertools
le = LabelEncoder()
le.fit(['fear','happy','neutral','anger','sad'])
le.inverse_transform([0,1,2,3,4])
save_confusion_matrix('Inception model',le.inverse_transform([0,1,2,3,4]), true_val, prediction)
```

```python
ctrue_val = np.load('charles_true.npy')
cprediction = np.load('charles_pred.npy')

rtrue_val = np.load('rafi_y_true.npy')
rprediction = np.load('rafi_y_pred.npy')
```

```python
matplotlib.rcParams.update({'font.size': '13'})
matplotlib.rcParams.update({'text.usetex': 'False'})
```

```python
plt.figure(figsize=(15,5))
plt.subplot(131)
save_confusion_matrix('Inception',le.inverse_transform([0,1,2,3,4]), true_val, prediction)
#plt.colorbar()
plt.tight_layout()
plt.subplot(132)
save_confusion_matrix('VGG + EmotiW',le.inverse_transform([0,1,2,3,4]), ctrue_val, cprediction)
plt.tight_layout()
plt.subplot(133)
save_confusion_matrix('CNN',le.inverse_transform([0,2,1,3,4]), rtrue_val, rprediction)
plt.tight_layout()
```
